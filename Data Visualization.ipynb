# Visualizing the dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from math import sqrt
import seaborn as sns
```

```python
train = pd.read_csv('fashion-mnist_train.csv', sep = ',')
```

```python
test = pd.read_csv('fashion-mnist_test.csv', sep=',')
```

```python
train.head()
```

```python
test.tail()
```

```python
ntrain = train.shape[0]
ntest = test.shape[0]

ytrain = train["label"]
xtrain = train.drop("label", axis=1)

dim = int(sqrt(xtrain.shape[1]))
```

```python
xtrain_vis = xtrain.values.reshape(ntrain, dim, dim)


for i in range(0,9): 
    plt.subplot(330 + (i+1)) 
    plt.imshow(xtrain_vis[i], cmap=plt.get_cmap('gray'))
    plt.title(ytrain[i]);
```
